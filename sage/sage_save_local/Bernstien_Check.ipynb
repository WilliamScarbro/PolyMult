```python

class Quotient_Ring:
    def __init__(self,n,p,N,d):
        ff.<gen>=GF(p,modulus="primitive") # coeficient field
        self.gen=gen
        self.ff=ff
        assert N%n==0, "{}|{} is false".format(n,N)
        self.p=p
        self.N=N
        self.n=n
        self.d=d
    def __str__(self):
        return "Quotient Ring: ({}) / (x^{}-(w_{})^{})".format(self.ff,self.n,self.N,self.d)
    # a and b are lists of ff of length n
    def multiply(self,a,b):
        pass
    def denominator(self):
        pr=ZZ["x"] # ZZ should be self.ff, but self.ff doesn't work with E
        x=pr("x")
        return x^self.n-E(self.N)^(self.d)
    def root(self,degree):
        assert (self.p-1)%degree==0, (self.p,degree)
        return self.gen^((self.p-1)//degree)
    def smush(self,a):
        print("smush")
        w=self.root(self.N)^(self.d)
        #print("w",w)
        k=len(a)//self.n
        new_a=[0 for i in range(self.n)]
        for i in range(k):
            for j in range(self.n):
                el=a[i*self.n+j]*w^i
                #print((i,j),el)
                new_a[j]+=el
        return new_a
    def gamma_sig(self):
        return Quotient_Ring(self.n,self.p,self.N,self.N)
    def gamma(self,a):
        assert len(a)==self.n
        assert (self.N-self.d)%self.n==0
        zeta=self.root(self.N)^((self.N-self.d)/self.n)
        return [a[i]*zeta^(i) for i in range(self.n)]
    def gamma_inv(self,a):
        assert len(a)==self.n
        assert (self.N-self.d)%self.n==0
        zeta=self.root(self.N)^((self.N-self.d)/self.n)
        return [a[i]*zeta^(self.N-i) for i in range(self.n)]
    
class QR_El:
    def __init__(self,qr,vec):
        self.qr=qr
        self.vec=vec
        for i in range(len(vec)):
            self.vec[i]*=self.qr.ff("1")
    def __getitem__(self,i):
        return self.vec[i]
    def __len__(self):
        return len(self.vec)
# eq
class Multiply:
    def __init__(self,ring):
        self.ring=ring
    def __str__(self):
        return "Multiply in "+str(self.ring)
    def __call__(self,a,b):
        assert a in self.ring
        assert b in self.ring
        return a*b

    
# reduction
class Bern_CT:
    def __init__(self,qr,k):
        assert qr.n%k==0
        self.qr=qr
        self.k=k
    def phi_sig(self):
        source=self.qr
        n=source.n//self.k
        dests=[]
        for i in range(self.k):
            d=source.d//self.k+i*source.N//self.k
            dests.append(Quotient_Ring(n,source.p,source.N,d))
        prod=1
        for dest in dests:
            prod*=dest.denominator()
        #print("product: {}".format(prod))
        return dests
    def phi(self,a):
        return [qr.smush(a) for qr in self.phi_sig()]
    def phi_inv(self,A):
        #print("A",A)
        m=self.qr.n//k
        a=[0 for i in range(self.qr.n)]
        w=self.qr.root(self.qr.N)
        d_k=self.qr.d//self.k
        M=self.qr.N//self.k
        #print("w",w)
        for i in range(k):
            for j in range(m):
                for z in range(k):
                    el=w^(-(d_k+z*M)*i)*A[z][j]
                    #print((i,j,z),el)
                    a[i*m+j]+=el
        for i in range(k):
            for j in range(m):
                a[i*m+j]/=k
        return a
```

```python
def poly_red(k):
    pr=ZZ['x','c']
    x=pr("x")
    c=pr("c")
    w=E(k)
    prod=1
    for i in range(k):
        prod*=(x-w^i*c)
    return prod
```
